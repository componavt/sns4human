```python
import pandas as pd
df = pd.read_csv('df_tokens.csv')
```

```python
dicts_lemm = {}
for indx, row in df.iterrows():
    post = row['tokens']
    date = row['year']
    if date not in dicts_lemm:
      dicts_lemm[date] = {}
    tokens = post.split()
    for word in tokens:
      if dicts_lemm[date].get(word, 0) == 0:
        dicts_lemm[date][word] = 1
      else:
        dicts_lemm[date][word] += 1
    dicts_lemm[date] = dict(sorted(dicts_lemm[date].items(), key=lambda item: item[1], reverse=True))
    dicts_lemm[date] = dict(list(dicts_lemm[date].items())[:10])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import imageio
import os

# Пример данных
data = {
    2024: {
        'карелия': 1141,
        'год': 572,
        'народ': 469,
        'вепсский': 165,
        'муниципальный': 141
    }
}

# Подготовка данных
years = sorted(data.keys())
words = sorted(set(word for year in data for word in data[year].keys()))

# Создание кадров для GIF
filenames = []

for year in years:
    frequencies = [data[year].get(word, 0) for word in words]

    plt.figure(figsize=(20, 18))
    bars = plt.bar(words, frequencies, color='skyblue')
    plt.title(f'Самые распространенные слова в {year}', fontsize=24)
    plt.xlabel('Слова', fontsize=20)
    plt.ylabel('Частота', fontsize=20)

    # Устанавливаем пределы по оси Y
    plt.ylim(0, max(max(frequencies), 1) + 50)

    # Устанавливаем порядок по оси X (годы)
    plt.xticks(rotation=45, fontsize=14)

    # Добавление значений частоты над столбцами
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom', fontsize=14)

    filename = f'{year}.png'
    plt.savefig(filename)
    plt.close()
    filenames.append(filename)

# Создание GIF с уменьшенной скоростью
with imageio.get_writer('histogram_animation.gif', mode='I', duration=1) as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)

# Удаление временных файлов
for filename in filenames:
    os.remove(filename)

print("GIF создан: histogram_animation.gif")
```
